### `poly_eval` test

```python
from internallib import poly_eval
import numpy as np

def main():
    import matplotlib.pyplot as plt
    # (Chebyshev polynomial of the first kind of degree 5)
    # T_5 = 16x^5 - 20x^3 + 5x 
    c = np.array([0., 5., 0., -20., 0., 16]) 
    
    x = np.linspace(-1, 1, 100)
    plt.plot(x, poly_eval(c, x))

if __name__ == "__main__":
    main()
```
